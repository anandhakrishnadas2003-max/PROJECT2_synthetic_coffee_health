1.Health and Lifestyle Analysis

*Creating dashboards and visual summaries of lifestyle and health patterns.

1.How to group by gender and get average coffee_Intake?

```python
df.groupby('Gender')['Coffee_Intake'].mean()
```

2.HOW to sort the dataset by age in descending order?

```python
df.sort_values('Age',ascending=False)
```

3.How to get the unique values in the country column

```python
df['Country'].unique()
```

4.How to replace 'fair' in  sleep_quality with 'average'


```python
df['Sleep_Quality']=df['Sleep_Quality'].replace('fair','average')
```

5.How to drop the ID column 

```python
df.drop('ID',axis=1,inplace=True)
```

5. How to calculate the correlation between Sleep_Hours and Coffee_Intake using numpy?


```python
np.mean(df['BMI']), np.std(df['BMI'])
```

6. How to normalize the Caffeine_mg column?

```python
df['Caffeine_mg_norm'] = (df['Caffeine_mg'] - df['Caffeine_mg'].min()) / (df['Caffeine_mg'].max() - df['Caffeine_mg'].min())
print(df['Caffeine_mg_norm'])
```

7. How to count how many entries are above average in Heart_Rate?


```python
np.sum(df['Heart_Rate'] > np.mean(df['Heart_Rate']))
```

8. How to create a numpy array from the Sleep_Hours column?


```python
sleep_hours_array = np.array(df['Sleep_Hours'])
print(sleep_hours_array)
```

8.How to plot a histogram of Coffee_Intake

```python
import matplotlib.pyplot as plt
plt.hist(df['Coffee_Intake'], bins=20)
plt.xlabel('Coffee Intake')
plt.ylabel('Frequency')
plt.title('Distribution of Coffee Intake')
plt.show()
```

9. How to plot a scatter plot between Caffeine_mg and Sleep_Hours

```python
plt.scatter(df['Caffeine_mg'], df['Sleep_Hours'])
plt.xlabel('Caffeine (mg)')
plt.ylabel('Sleep Hours')
plt.title('Caffeine vs Sleep Hours')
plt.show()
```

10.How to plot the average Sleep_Hours for each Sleep_Quality?


```python

df.groupby('Sleep_Quality')['Sleep_Hours'].mean().plot(kind='bar')
plt.title('Average Sleep Hours by Sleep Quality')
plt.show()
```

11. How to plot a line graph of Age vs Heart_Rate?


```python
df.plot(x='Age', y='Heart_Rate', kind='line')
plt.title('Heart Rate over Age')
plt.show()
```

12. How to create a boxplot of BMI grouped by Sleep_Qualit


```python
import seaborn as sns
sns.boxplot(x='Sleep_Quality', y='BMI', data=df)
plt.title('BMI by Sleep Quality')
plt.show()
```

13.How to create a count plot of Stress_Level


```python
sns.countplot(x='Stress_Level', data=df)
plt.title('Distribution of Stress Levels')
plt.show()
```

14.How to create a heatmap of correlation between numerical variables

```python
corr = df.corr()
sns.heatmap(corr, annot=True, fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

15.How to create a violin plot of Physical_Activity_Hours grouped by Health_Issues

```python
sns.violinplot(x='Health_Issues', y='Physical_Activity_Hours', data=df)
plt.title('Activity Hours by Health Issues')
plt.show()
```


16.How to create a pair plot of Coffee_Intake, Caffeine_mg, and Sleep_Hours


```python

sns.pairplot(df[['Coffee_Intake', 'Caffeine_mg', 'Sleep_Hours']])
plt.show()
```

17.How to style plots with seaborn themes

```python
sns.set_style('whitegrid')
sns.histplot(df['Coffee_Intake'])
plt.title('Coffee Intake with White Grid')
plt.show()
```

Health:

. Average BMI ≈ 24 (healthy range), though a portion is overweight.

. Heart rates are mostly normal (mean ≈ 71 bpm).

. Nearly 60% report no health issues, but some have cardiovascular or metabolic conditions.

Lifestyle:

. Physical activity averages 7.5 hrs/week, but some are inactive.

. About 20% are smokers, and 30% consume alcohol.

. Occupations are varied, with office jobs being most common.
